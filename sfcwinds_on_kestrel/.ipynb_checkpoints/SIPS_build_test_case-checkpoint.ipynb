```python
# Load Observation file NM (legacy)

# obs= pd.read_pickle(f"/kfs2/projects/sfcwinds/obs_AZ_2023.pkl") # this is the assembled files from Bilal
obs= pd.read_pickle(f"/kfs2/projects/sfcwinds/observations_intermediate_steps/obs_NM_synoptic_2024.pkl") # this is the 2024 1-year data from synoptic

# obs['windspeed'] = obs['windspeed'].replace(0, np.nan)
# obs['gust'] = obs['gust'].replace(0, np.nan)
# obs.replace('', np.nan, inplace=True)

obs = obs.apply(pd.to_numeric, errors="ignore")

obs.index = pd.to_datetime(obs.index)
obs.index = obs.index.tz_localize(None)

obs_old = obs

# np.unique(obs.station_name)
```

```python
# Read metadata to filter what we want
meta = pd.read_csv("/kfs2/projects/sfcwinds/observations/metadata_CONUS.csv")

# filter which data to read
base_dir = "/kfs2/projects/sfcwinds/observations/"
stations = meta[meta.state == "NM"].station_id.values
networks = ["*"]
years    = [date_string[:4]]

#Read observations according to filter
years = [str(y) for y in years]
all_dfs = []

for net in networks:
    for station in stations:
        for year in years:
            pattern = os.path.join(base_dir, net, station, f"{year}.parquet")
            found = glob.glob(pattern)

            for f in found:
                try:
                    df = pd.read_parquet(f)
                    # Extract station_id from the path (e.g., .../CoAgMet/HOT01/2023.parquet)
                    station_id = os.path.basename(os.path.dirname(f))  # 'HOT01'
                    df["station_id"] = station_id
                    all_dfs.append(df)
                except Exception as e:
                    print(f"Failed to read {f}: {e}")

if not all_dfs:
    print("No files found.")
else:
    print(f"{len(all_dfs)} files found.")
    obs = pd.concat(all_dfs, ignore_index=True)
    obs["timestamp"] = obs["timestamp"].dt.tz_convert(None)

# Merge with metadata
obs = obs.merge(meta[["station_id", "lat", "lon", "height", "elev", "source_network", "state"]], on="station_id", how="left")
```

```python
# Load HRRR files

HRRR_folder = "/kfs2/projects/sfcwinds/HRRR"
area = "US_SW"


file_pattern = os.path.join(HRRR_folder, f"hrrr_{area}_{date_string[:]}*.nc")
file_list = sorted(glob.glob(file_pattern))
if file_list:
    hrrr = xr.open_mfdataset(file_list[:], concat_dim ="valid_time",combine='nested', chunks="auto", parallel = True)

hrrr['longitude'] = xr.where(hrrr['longitude'] > 180, hrrr['longitude'] - 360, hrrr['longitude'])

# test hourly vs subourly wind speed
# plt.figure()
# plt.plot(hrrr.valid_time.values, hrrr.isel(x=100, y=100).u10_h.values, ".", color="blue", label = "HRRR sfc")    
# plt.plot(hrrr.valid_time.values, hrrr.isel(x=100, y=100).u10.values, ".", color="orange", label = "HRRR subh") 
# plt.plot(hrrr.valid_time.values, hrrr.isel(x=100, y=100).v10_h.values, "x", color="blue", label = "HRRR sfc")    
# plt.plot(hrrr.valid_time.values, hrrr.isel(x=100, y=100).v10.values, "x", color="orange", label = "HRRR subh") 

hrrr["wspd10"] = (hrrr.u10**2 + hrrr.v10**2)**0.5
```

```python
# Load ERA5 files

# Define source
source = "downloaded"

if source == "sup3rwind":
    ERA5_folder = "/kfs2/datasets/sup3rwind/era5/conus_large"
    file_pattern = os.path.join(ERA5_folder, f"era5_conus_{date_string[:4]}.nc")
elif source == "downloaded":
    ERA5_folder = "/kfs2/projects/sfcwinds/ERA5"
    file_pattern = os.path.join(ERA5_folder, f"ERA5_surface_{date_string[:].replace("-", "_")}*.nc")

# List of all files
file_list = sorted(glob.glob(file_pattern))

if file_list:
    
    # Read files in list
    era5 = xr.open_mfdataset(file_list[:], concat_dim ="valid_time",combine='nested', parallel = True)

    # Drop coordinates that have only one coordinate 
    # ERA5T data (temporary): expver=0005, ERA5 data (final): expver=0001.
    # number: ensemble member ( only one with deterministic forecast?)
    era5 = era5.drop_vars(['expver', 'number'])#.drop_dims(["valid_time"])

    if source == "sup3rwind":
        # Make latitude and longitude dimensions instead of west_east and south_north, which are just running numbers from 0 to max
        era5 = era5.assign_coords(   
            latitude =("south_north", np.unique(era5.latitude)),
            longitude =("west_east", np.unique(era5.longitude))
        )
        era5 = era5.swap_dims({"south_north": "latitude", "west_east": "longitude"})
    
        # check that all latitudes are the same for one south_north value:
        # for i in range(0,len(era5.south_north)):
        #     print (np.all(era5.isel(south_north=i).latitude.values))

else:
    era5 = xr.Dataset()
    print ("No ERA5 data found.")

# era5 = era5.assign_coords(
#     longitude=((era5.longitude + 180) % 360) - 180
# )

era5["wspd10"] = (era5.u10**2 + era5.v10**2)**0.5
```

```python
#%% To do: combine HRRR and ERA5 grid

# 1) Re-grid HRRR

# Extract the 2D latitude and longitude values for the selected valid time
lat2d = hrrr.latitude.isel(valid_time=0)
lon2d = hrrr.longitude.isel(valid_time=0)

# Stack y and x dimensions into a single 'points' dimension
hrrr_flat = hrrr.stack(points=('y', 'x'))

# Add lat and lon as coordinates on the new 'points' dimension
hrrr_flat = hrrr_flat.assign_coords({
    'latitude': ('points', lat2d.values.ravel()),  # Flatten latitude
    'longitude': ('points', lon2d.values.ravel())   # Flatten longitude
})

# Drop the original x and y dimensions as they are no longer needed
hrrr_flat = hrrr_flat.drop_vars(['y', 'x', 'points'])

```

```python
# 2) Add ERA5 data to regridded HRRR

# Sarch for the closest lat, lon in hrrr_flat for each ERA5 grid point

# Extract latitude and longitude from hrrr_flat (flattened grid)
hrrr_lat = hrrr_flat.latitude.values
hrrr_lon = hrrr_flat.longitude.values

# Create an empty array to store the indices of closest points
closest_indices = []

# Loop over all ERA5 points (assuming they are 2D arrays)
for era5_lat_val, era5_lon_val in zip(era5.latitude, era5.longitude):
    # Calculate distances from this ERA5 point to all points in hrrr_flat
    distances = np.array([geodesic((era5_lat_val, era5_lon_val), (lat, lon)).meters
                         for lat, lon in zip(hrrr_lat, hrrr_lon)])
    
    # Find the index of the closest point in hrrr_flat
    closest_index = np.argmin(distances)
    closest_indices.append(closest_index)

# Convert the closest_indices list to a numpy array
closest_indices = np.array(closest_indices)

# Now we can assign ERA5 data to hrrr_flat based on closest matches
# For example, let's say we want to add a specific variable from ERA5 (e.g., temperature)

# Assuming you have an ERA5 variable like `era5_temperature`
era5_temperature = era5_variables.temperature.values.flatten()  # Flatten the ERA5 temperature values

# Assign ERA5 temperature values to hrrr_flat points dimension based on closest indices
hrrr_flat['era5_temperature'] = xr.DataArray(
    era5_temperature[closest_indices],
    coords={'points': hrrr_flat.points},
    dims=['points']
)

# Print the updated hrrr_flat dataset with the ERA5 temperature variable
print(hrrr_flat)
```
